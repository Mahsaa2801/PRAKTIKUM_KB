## 3. Tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
data = {}
data["Jumlah Record"] = df.shape[0]
data["Jumlah Attribute"] = df.shape[1]
data["Nama Attribute"] = df.columns.tolist()
data["Jumlah Attribute Dengan Tipe Angka"] = df.select_dtypes(include=["int64", "float64"]).shape[1]

dictKeys = list(data.keys())

table = prtl.PrettyTable()
table.add_column(dictKeys[0], [data["Jumlah Record"]])
table.add_column(dictKeys[1], [data["Jumlah Attribute"]])
table.add_column(dictKeys[2], ['\n'.join(data['Nama Attribute'])])
table.add_column(dictKeys[3], [data["Jumlah Attribute Dengan Tipe Angka"]])
print(table)
```
