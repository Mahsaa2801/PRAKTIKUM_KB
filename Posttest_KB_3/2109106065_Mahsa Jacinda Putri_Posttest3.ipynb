### Import Library

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Import Dataset

```python
df = pd.read_csv('Downloads/master.csv')
```

### Melihat Dataset

```python
df
```

### GRAFIK BATANG

```python
generasi = df["generation"].value_counts()
plt.bar(generasi.index, generasi, color = ["darkblue", "darkgreen","red","yellow","magenta","lightblue"])
plt.xlabel("Generasi", fontweight = "bold")
plt.ylabel("Jumlah", fontweight = "bold")
plt.title("Grafik Perbandingan Generasi", fontsize = 15, fontweight = "bold")
plt.xticks(rotation=60, ha='right')
plt.show()
```

```python
generasi
```

#### Dari grafik batang di atas dapat di ketahui:
#### - Kasus bunuh diri paling banyak di lakukan oleh Generation X (individu yang lahir di antara tahun 1965-1976) sebanyak 6408 kasus
#### - Kasus bunuh diri terendah ada di Generation Z (individu yang lahir di tahun 1995-2012) yaitu sebanyak 1470 kasus

### Heatmap

```python
sns.heatmap(df.corr(),
            annot = True)
plt.title("Heatmap", fontsize = 20)
plt.show()
```

### Scatter Plot

```python
df.plot.scatter("suicides_no", "year")
plt.title('Scatter Plot Relasi Antara Jumlah Bunuh diri dan Tahun terjadinya') 
print(df.corr())
```

#### Dari grafik di atas dapat diketahui bahwa pada rentang tahun 1990 - 2005 terjadi peningkatan kasus bunuh diri

### Pie Chart

```python
labels = ['Male', 'Female']
plt.pie(df['sex'].value_counts(), labels=labels, autopct="%1.1f%%") 
plt.title('Jenis Kelamin') 
plt.show() 
```

#### Dari grafik di atas di ketahui jumlah pria dan wanita dalam kasus bunuh diri adalah sama
